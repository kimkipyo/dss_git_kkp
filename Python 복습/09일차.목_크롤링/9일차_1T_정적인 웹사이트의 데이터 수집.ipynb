```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("http://www.fastcampus.co.kr/courses/")
```

```python
response.status_code, response
```

```python
dom = BeautifulSoup(response.text, "html.parser")
```

```python
# 그룹 ( Language, Programming, Business, ... )
# 강의명 ( Node.js 로 시작하는 웹 프로그래밍 CAMP, ... )
# 강의 링크 ( http://fastcampus.co.kr/dev_camp_nodejs, ... )
```

```python
dom.select_one("#mk-page-id-26880")
```

```python
dom.select_one("#mk-page-id-26880").select("div.wpb_row.vc_row")
```

```python
camp_team_elements = dom.select_one("#mk-page-id-26880").select("div.wpb_row.vc_row")
```

```python
len(camp_team_elements)
```

```python
camp_team_elements[0]  #~[2], [3] ... 하나씩 찾아보기
```

```python
camp_team_elements[5]
```

```python
camp_team_elements = camp_team_elements[5::2]   #나중에 중복 문제 때문에 ::2 두 칸 건너 뛰어 중복 제거
```

```python
camp_team_element = camp_team_elements[0]
```

```python
camp_team_element.select_one("h2.vc_custom_heading").text
```

```python
for camp_team_element in camp_team_elements:
    print(camp_team_element.select_one("h2.vc_custom_heading").text)
```

```python
course_elements = camp_team_element.select("a.vc_general")
```

```python
len(course_elements)
```

```python
for course_element in course_elements:
    title = course_element.text
    url = course_element.get("href")
    print(title)
    print(url)
```

```python
for camp_team_element in camp_team_elements:
    team_name = camp_team_element.select_one("h2.vc_custom_heading").text
    
    for course_element in course_elements:
        title = course_element.text
        url = course_element.get("href")
        
        print((team_name, title, url))
```
