# 1T_피보나치 구현 - Class 를 이용하여 구조화해서 해결하기

```python
#숙제 1번
chache = {} # key, value pair 를 사용하기 위해서 dict에 데이터 저장
class Fibonacci():

    def __init__(self):
        self.cache = {}    # 빈 딕셔너리다.
    
    def __call__(self, n):
        if n in self.cache:   
            return self.cache[n]
        # n이 self.cache에 없는 경우 -> n<0, n=0 n=1 -> fibo(0)=0, fibo(1)=1
        if n < 0:
            return 0
        if n == 0:
            self.cache[n] = 0
            return self.cache[n]   # return fibo(0)
        if n == 1:
            self.cache[n] = 1
            return self.cache[n]    # return fibo(1)
        
        result = self.__call__(n-1) + self.__call__(n-2)
        self.cache[n] = result
        return self.cache[n]
```

```python
fibo = Fibonacci()    #피보나치 객체, 현재까지 저장된 피보 결과값 ({})
```

```python
fibo(7)
```

```python
#리펙토링
chache = {}

class Fibonacci():
    # 앞의 것은 정의가 재귀함수로 되어 있고 기본값으로 되어 있다.
    # 초기값 설정으로 날릴 수 있어
    def __init__(self):
        self.cache = {}
        self.cache[0] = 0
        self.cache[1] = 1
        
    def __call__(self, n):
        if n in self.cache:
            return self.cache[n]
        if n < 0:
            return 0
        
        result = self.__call__(n-1) + self.__call__(n-2)
        self.cache[n] = result
        return self.cache[n]
    
    def __str__(self):    #원하는 형태로 출력하기 위해서
        return "\n".join([      #한줄 한줄은 문자열의 리스트로 조인하겠다.
                "fibo({n}) = {result}".format(n=key, result=value)
                for key, value
                in self.cache.items()
            ])
```

```python
fibo = Fibonacci()    #피보나치 객체, 현재까지 저장된 피보 결과값 ({})
fibo(5)
```

# 3T_네이버 블로그 크롤링 - requests, beautifulsoup

```python
import requests
```

```python
# import bs4
from bs4 import BeautifulSoup
```

```python
BASE_URL = "https://search.naver.com/search.naver?where=post&sm=tab_jum&ie=utf8&query=%EC%8B%AC%ED%95%B4%EC%9B%90+%EC%86%94%EB%A1%9C%EC%9B%8C%EC%A6%88"
```

```python
# 1. requests => Crawling 하는 것이다.
#  Requests: HTTP for Humans _ 사람이 이해하기 쉽도록 쓰인 라이브러리
response = requests.get(BASE_URL)  #get이라는 함수를 실행 시킨 것이다.
```

```python
response.status_code   # 2XX => Successful ( 200 OK )
                       # 3XX => Redirection
                       # 4XX => Failed ( 403 Forbidden )
```

```python
response.content
```

```python
response.text
```

```python
dom = BeautifulSoup(response.content, "html.parser")
# dom ( 돔 ) => Document Object Model
# html > div > h1 > p > ul > li > ... 이거 하나하나를 객체로 보고
#그것들의 상관관계(속해 있다 라는 개념)를 구조화해서 보여준다.
dom
```

```python
# Parsing

# tag => "태그이름" ( "div", "a" )
# id => 문서 전체에서 unique 한 값 => "#" (하나만 뽑을 때 유용할 수도 있다.)
# class => 문서 전체에서 1~n => 공통적인 특징을 가진 tag 들 => "."
```

```python
post_elements = dom.select("li.sh_blog_top")
```

```python
len(post_elements)
```

```python
post_element = post_elements[0]
post_element
```

```python
post_element.select_one("a.sh_blog_title")   #type이 list이다.
```

```python
title_element = post_element.select_one("a.sh_blog_title")   #하나만 뽑으려면 _one을 붙이면 된다.
title_element
```

```python
for post_element in post_elements:
    title_element = post_element.select_one("a.sh_blog_title")
    title = title_element.text
    url = title_element.get("href")
    print((title, url))
```
